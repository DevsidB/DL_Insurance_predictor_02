## Complete Manual Mode on- Finding slopes and intercept with Gradient descent. External modules not allowed.

__Remember: Functions like log_loss mostly needs numpy arrays, as some summation occurs in every error functions(SSE,MSE,LogLoss). The functions which takes summation works mostly with the loops and undergoes an iterative process. So for error/cost/loss function make sure to provide numpy arrays as an input to the function.__

```python
def log_loss(y_true, y_predicted): # Made with precision to take numpy array as an input.
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted ]
    y_predicted_new = [min (i, 1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new) 
    return -np.mean(y_true* np.log(y_predicted_new)+ (1-y_true)*np.log(1-y_predicted_new))
```

We created a sigmoid function above which accepts single value of x. But while dealing with big datasets we need to deal with multiple values of x at once to provide as an input to the function. This makes the process much faster. So creating a sigmoid function below to accept numpy array as an inpuit instead of a single value.

```python
def sigmoid_python(X): # Made with precision to take numpy array as an input.
    return (1/ (1 + np.exp(-X)))
```

```python
sigmoid_python(np.array([1,12,20])) # just for checking
```

__Here comes the king - Gradient Descent. Lets find slopes and intercepts. Manual mode on.__

```python
def gradient_descent(x1,x2,y_true,epochs): #Gradient descent function for logistic regression with 2 weights.Generic implementation not done.
    w1=w2=1 #Taking 1 . Can also take 0
    bias = 0
    rate = 0.5 #usually 0.01
    n = len(x1)
    
    for i in range(epochs):
        weighted_sum = w1*x1 + w2*x2 +bias          # finding weighted sum
        y_predicted = sigmoid_python(weighted_sum)  # substituting in sigmoid
        loss = log_loss(y_true, y_predicted)        # calculating the loss

        w1d= (1/n)*sum(x1*(y_predicted- y_true))   ## Simplest way to find partial derivative
        w2d= (1/n)*sum(x2*(y_predicted- y_true))   ## Simplest way to find partial derivative 
        bd= np.mean(y_predicted- y_true) 
        
#         w1d = (1/n)*np.dot(np.transpose(x1),(y_predicted- y_true))  ## Just another way to find the partial derivative using numpy
#         w2d = (1/n)*np.dot(np.transpose(x2),(y_predicted- y_true))  ## Just another way to find the partial derivative using numpy
        
#         w1d = (1/n)*(y_predicted- y_true).dot(x1.transpose())  ## And one moreway to find the partial derivative using numpy.
#         w2d = (1/n)*(y_predicted- y_true).dot(x2.transpose())  ## And one moreway to find the partial derivative using numpy.
          
    
        w1 = w1 - rate*w1d
        w2 = w2 - rate*w2d
        bias = bias - rate*bd
                       
        print (f"epoch : {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}")
#         print(y_predicted) 
    return w1,w2,bias
```

Generally we stop the iterations where the loss is not changing for some consecutive iterations. Here we will stop the iterations when loss is equal to that of tensorflows output. Trying to replicate the weights that tensorflow showed up. We can achieve that by using a threshold. Threshold is a value which we provide to our function to break the loop once that value is reached. From line 12 --loss = 0.4631. 

__Using threshold__

```python
def gradient_descent(x1,x2,y_true,epochs,loss_threshold): #Gradient descent function for logistic regression.
    w1=w2=1 #Taking 1 . Can also take 0
    bias = 0
    rate = 0.5 #usually 0.01
    n = len(x1)
    
    for i in range(epochs):
        weighted_sum = w1*x1 + w2*x2 +bias          # finding weighted sum
        y_predicted = sigmoid_python(weighted_sum)  # substituting in sigmoid
        loss = log_loss(y_true, y_predicted)        # calculating the loss

        w1d= (1/n)*sum(x1*(y_predicted- y_true))   ## Simplest way to find partial derivative
        w2d= (1/n)*sum(x2*(y_predicted- y_true))   ## Simplest way to find partial derivative 
        bd= np.mean(y_predicted- y_true) 
        
#         w1d = (1/n)*np.dot(np.transpose(x1),(y_predicted- y_true))  ## Just another way to find the partial derivative using numpy
#         w2d = (1/n)*np.dot(np.transpose(x2),(y_predicted- y_true))  ## Just another way to find the partial derivative using numpy
        
#         w1d = (1/n)*(y_predicted- y_true).dot(x1.transpose())  ## And one moreway to find the partial derivative using numpy.
#         w2d = (1/n)*(y_predicted- y_true).dot(x2.transpose())  ## And one moreway to find the partial derivative using numpy.
          
    
        w1 = w1 - rate*w1d
        w2 = w2 - rate*w2d
        bias = bias - rate*bd
        
        if loss<=loss_threshold:
            break
                       
        print (f"epoch : {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}")
#         print(y_predicted) 
    return w1,w2,bias
```
